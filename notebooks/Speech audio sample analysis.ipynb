```python
import glob
import IPython

from pydub import AudioSegment
```

## No Speech Sample

```python
no_speech_filename = 'audio_examples/no_speech_sample_2s.wav'
audio = AudioSegment.from_wav(no_speech_filename)

print('no speech')
print('max db = {}'.format(audio.max_dBFS))

IPython.display.Audio(no_speech_filename)
```

## Speech Sample

```python
speech_filename = 'audio_examples/speech_sample_2s.wav'
audio = AudioSegment.from_wav(speech_filename)

print('speech')
print('max db = {}'.format(audio.max_dBFS))

IPython.display.Audio(speech_filename)
```

---

```python
speech_wav_files = glob.glob('../data/dev_set_wav/dialog/*.wav')
print(len(speech_wav_files))
```

## List the silent audio sample

```python
silence_filenames = []
for speech_wav_file in speech_wav_files:
    audio = AudioSegment.from_wav(speech_wav_file)

    #if -15 < audio.max_dBFS < -14:
    if audio.max_dBFS < -15:
        silence_filenames.append(speech_wav_file)
    
print('number of silence sample = {}/{}'.format(len(silence_filenames), len(speech_wav_files)))
```

---

## Listen to silent samples

```python
i = 0
```

```python
if i < len(silence_filenames):
    i += 1

print('sample {} - {}'.format(i-1, silence_filenames[i-1]))
IPython.display.Audio(silence_filenames[i-1])
```
